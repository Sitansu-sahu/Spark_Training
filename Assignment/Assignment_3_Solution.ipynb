```
# a)	Create HR_Employee Pysaprk dataFrame. 
HR_Employee = spark.read.csv("dbfs:/FileStore/tables/HR_Employee.csv",header = True , inferSchema = True)
```

```
#b)	Show printSchema & show 10 rows of Dataset
HR_Employee.show(10,False)
```

```
HR_Employee.createOrReplaceTempView("HR_Emp_df")
```

```
# c)	Most employee is working in which department
spark.sql("select Department ,count(*) as most_nu_employee from HR_Emp_df group by Department order by most_nu_employee desc ").show()
```

```
# d)	Highest number of job roles
spark.sql("select Max(JobRole) as Job_role,count(*) as most_nu_employee from HR_Emp_df  ").show()
```

```
# e)	Which gender have higher strength as workforce?

spark.sql("select Max(Gender) as Gender,count(*) as most_Strength from HR_Emp_df  ").show()
```

```
# f)	Compare the marital status of employee and find the most frequent status.
spark.sql("select MaritalStatus as Gender,count(*) as most_frequent_status from HR_Emp_df where BusinessTravel ='Travel_Frequently' group by MaritalStatus ").show()
```

```
# g)	Mostly hired employee have qualification

spark.sql("select Education, count(*) as Number_Employee from HR_Emp_df  group by Education order by Number_Employee desc").show()
```

```
# h)	Find the count of employee from which education fields 

spark.sql("select EducationField, count(*) as Number_Employee from HR_Emp_df  group by EducationField order by Number_Employee desc").show()
```

```
# i)	What is the job satisfaction level of employee?
spark.sql("select    JobSatisfaction , count(*) as Number_Employee from HR_Emp_df  group by JobSatisfaction order by Number_Employee desc").show()
```

```
# j)	Does most of employee do overtime: Yes or No?

spark.sql("select Overtime, count(*) as Number_Employee from HR_Emp_df where OverTime ='Yes' or Overtime ='No' group by Overtime ").show()
```

```
# k)	Find Min & Max Salaried employees. 
spark.sql("select Max(Income) as Maximun_sal,Min(Income) as Minimun_sal  from HR_Emp_df ").show()
```

```
# l)	Does most of the employee do business travel? Find of the employees counts for each category   BusinessTravel   
spark.sql("select JobRole , count(*) as Number_Employee from HR_Emp_df where BusinessTravel ='Travel_Frequently' group by JobRole ").show()
```

```
# m)	Find the AVG salary of graduate employee.

spark.sql("select mean(Income) as Averge_sal_employee  from HR_Emp_df where Education = 'Bachelor' ").show()
```

```
#n)    Find the employee qualification receiving salary lower than equal to avg. salary of all employee. 

spark.sql("select Income, Education from HR_Emp_df where Income <= (select avg(Income) from HR_Emp_df)").show()

```

```
#o)    When does the employee have highest chance of promotion in terms of working year
spark.sql("select Performance_Rating, count(*) as PromotionChance from HR_Emp_df  group by Performance_Rating order by PromotionChance desc ").show()
```

```
#p)    Highest attrition is in which department?
spark.sql("select Department, count(*) as totalcount from HR_Emp_df where Attrition='Yes' group by Department order by totalcount desc ").show()
```

```
#q)    Show marital status of Person having highest attrition rate

spark.sql("select MaritalStatus, count(*) as HighAttrition from HR_Emp_df where Attrition='Yes' group by MaritalStatus order by HighAttrition desc ").show(1)
```
